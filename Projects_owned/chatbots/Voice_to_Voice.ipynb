```python
import speech_recognition as sr

def get_voice_input():
    """Captures voice input from the microphone and converts it to text."""
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening...")
        recognizer.adjust_for_ambient_noise(source) # Adjust for background noise
        audio = recognizer.listen(source,phrase_time_limit=10000)

    try:
        print("Recognizing...")
        text = recognizer.recognize_google(audio) # Use Google Web Speech API
        print(f"You said: {text}")
        return text
    except sr.UnknownValueError:
        print("Could not understand audio")
        return None
    except sr.RequestError as e:
        print(f"Could not request results from Speech Recognition service; {e}")
        return None
```

```python
import pyttsx3

def speak_text(text):
    """Converts text to speech and speaks it out."""
    engine = pyttsx3.init()
    engine.say(text)
    engine.runAndWait()
```

```python
import google.generativeai as genai
def generate_gemini_response(conversation_history, user_text_input): # Modified to accept conversation history
    """
    Sends user input and conversation history to Gemini API and gets the response.

    Args:
        conversation_history (list): A list of strings representing the conversation history so far.
        user_text_input (str): The text input from the user.

    Returns:
        str: The text response generated by Gemini, or None if there's an error.
    """
    system_prompt = """You are a helpful AI assistant designed to book doctor's appointments.
        Your goal is to understand the user's request and book an appointment for them.
        You can ask clarifying questions to get the necessary information, such as:
        - What type of doctor are you looking for (specialty)?
        - What is your preferred date or time for the appointment?
        - Do you have a preferred location?

        Keep the conversation natural and helpful. Once you have enough information to book,
        you can say something like: "Okay, let me check availability for you." (For this example, we are not connecting to a real API, so just simulate the booking process and confirm).

        Respond in a concise and helpful way. Remember the previous turns of the conversation to maintain context.
        """

    # Construct the full prompt by combining system prompt, history, and current user input
    prompt_parts = [system_prompt] + conversation_history + [user_text_input]

    GOOGLE_API_KEY = "AIzaSyAGHkaRYf8Hv02nXOFOvx3yZ4clkqL1IXs"  # **IMPORTANT: Replace with your actual API key**
    genai.configure(api_key=GOOGLE_API_KEY)

    # Select the Gemini model you want to use (e.g., 'gemini-pro')
    model = genai.GenerativeModel('gemini-2.0-flash')
    try:
        response = model.generate_content(prompt_parts)
        return response.text
    except Exception as e:
        print(f"Error generating Gemini response: {e}")
        return "Sorry, I encountered an error while processing your request."


def run_voice_appointment_agent_gemini_context(): # Renamed function to indicate context handling
    print("Doctor's Appointment Booking Agent (Voice-Based - Gemini - Context Aware)")
    speak_text("Hello, how can I help you today?")

    conversation_history = [] # Initialize an empty list to store conversation history

    while True:
        print("Listening for your request...")
        voice_input_text = get_voice_input()
        if not voice_input_text:
            speak_text("Sorry, I didn't catch that. Could you please repeat?")
            continue

        user_input = voice_input_text.lower()

        if user_input == "exit":
            speak_text("Goodbye!")
            break

        # --- Gemini API Integration with Context ---
        agent_response_text = generate_gemini_response(conversation_history, user_input) # Pass history to Gemini

        if agent_response_text:
            print(f"Agent Response (Gemini): {agent_response_text}")
            speak_text(agent_response_text)

            # Update conversation history: add user input and agent response
            conversation_history.append(user_input) # Add user input to history
            conversation_history.append(agent_response_text) # Add agent response to history
        else:
            speak_text("Sorry, I couldn't generate a response at this time.")


```
